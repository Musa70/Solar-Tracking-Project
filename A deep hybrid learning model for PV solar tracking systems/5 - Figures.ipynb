```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
mae = pd.read_csv('/Users/musaphiri/Documents/Solar Tracking Research/Solar tracking dataset document/Solar tracking notebooks/csv_metrics/MAE.csv')
mape = pd.read_csv('/Users/musaphiri/Documents/Solar Tracking Research/Solar tracking dataset document/Solar tracking notebooks/csv_metrics/MAPE.csv')
rmse = pd.read_csv('/Users/musaphiri/Documents/Solar Tracking Research/Solar tracking dataset document/Solar tracking notebooks/csv_metrics/RMSE.csv')
```

## Bar Chart

```python
# calculate the average of each column
mae_avg = mae.mean()
mape_avg = mape.mean()
rmse_avg = rmse.mean()
```

##### MAE_bar chart

```python
# create a new figure with a larger size and white background
fig = plt.figure(figsize=(13, 7), facecolor='w')

# create a bar plot of the average values
plt.bar(mae_avg.index, mae_avg.values, width=0.4, align='center', edgecolor='black')

# set the labels for the x and y axes
plt.xlabel('Models', fontsize=18)
plt.ylabel('average_mae_score', fontsize=18)

plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)

# set the title of the plot
#plt.title('Average Values of Columns')

# save the plot as a PNG file
plt.savefig('/Users/musaphiri/Documents/Solar Tracking Research/Solar tracking dataset document/Solar tracking notebooks/Figures/average_mae1.png')

# show the plot
plt.show()
```

## BOX PLOT

##### MAE_boxplot

```python
# create a list of arrays, where each array contains the data for a particular column
data = [mae['LSTM'], mae['CNN'], mae['SCT-CNN-LSTM'], mae['GRU']]

# set the colors for the boxes, whiskers, and medians
box_colors = ['#6495ED', '#F08080', '#90EE90', '#FFA07A', '#ADD8E6']
# whisker_colors = ['#2E8B57', '#8B0000', '#006400', '#A52A2A', '#0000FF']
median_colors = ['#000000', '#000000', '#000000', '#000000', '#000000']

# create a new figure with a larger size and white background
fig = plt.figure(figsize=(13, 7), facecolor='w')

# create a box plot of the data
bp = plt.boxplot(data, patch_artist=True, labels=['LSTM', 'CNN', 'SCT-CNN-LSTM', 'GRU'])

# customize the colors of the boxes, whiskers, and medians
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)
# for whisker, color in zip(bp['whiskers'], whisker_colors):
#     whisker.set_color(color)
for median, color in zip(bp['medians'], median_colors):
    median.set_color(color)

# set the labels for the x and y axes
plt.xlabel('models', fontsize=18)
plt.ylabel('mae_score', fontsize=18)

plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)

# add a legend
# boxes = [plt.Rectangle((0,0),1,1, color=c) for c in box_colors]
# whiskers = [plt.Line2D([],[], color=c) for c in whisker_colors]
# medians = [plt.Line2D([],[], color=c) for c in median_colors]
# plt.legend(boxes+whiskers+medians, ['Box', 'Whisker', 'Median'], loc='upper right')

# set the title of the plot
#plt.title('Box Plot of Four Columns', fontsize=16)

# remove grid lines
plt.grid(False)

# save the plot as a PNG file
plt.savefig('/Users/musaphiri/Documents/Solar Tracking Research/Solar tracking dataset document/Solar tracking notebooks/Figures/mae_boxplot.png')

# show the plot
plt.show()
```

```python
# Calculate and print the values for each box plot
q1, median, q3 = mae['LSTM'].quantile([0.25, 0.5, 0.75])
iqr = q3 - q1
upper_whisker = q3 + 1.5 * iqr
lower_whisker = q1 - 1.5 * iqr
print("Box plot for", 'NN')
print("Median: ", median)
print("Lower Quartile (Q1): ", q1)
print("Upper Quartile (Q3): ", q3)
print("Interquartile Range (IQR): ", iqr)
print("Upper Whisker: ", upper_whisker)
print("Lower Whisker: ", lower_whisker)

print(" ")

# Calculate and print the values for each box plot
q1, median, q3 = mae['CNN'].quantile([0.25, 0.5, 0.75])
iqr = q3 - q1
upper_whisker = q3 + 1.5 * iqr
lower_whisker = q1 - 1.5 * iqr
print("Box plot for", 'MM')
print("Median: ", median)
print("Lower Quartile (Q1): ", q1)
print("Upper Quartile (Q3): ", q3)
print("Interquartile Range (IQR): ", iqr)
print("Upper Whisker: ", upper_whisker)
print("Lower Whisker: ", lower_whisker)

print(" ")

# Calculate and print the values for each box plot
q1, median, q3 = mae['SCT-CNN-LSTM'].quantile([0.25, 0.5, 0.75])
iqr = q3 - q1
upper_whisker = q3 + 1.5 * iqr
lower_whisker = q1 - 1.5 * iqr
print("Box plot for", 'MD')
print("Median: ", median)
print("Lower Quartile (Q1): ", q1)
print("Upper Quartile (Q3): ", q3)
print("Interquartile Range (IQR): ", iqr)
print("Upper Whisker: ", upper_whisker)
print("Lower Whisker: ", lower_whisker)

print(" ")

# Calculate and print the values for each box plot
q1, median, q3 = mae['GRU'].quantile([0.25, 0.5, 0.75])
iqr = q3 - q1
upper_whisker = q3 + 1.5 * iqr
lower_whisker = q1 - 1.5 * iqr
print("Box plot for", 'MMAD')
print("Median: ", median)
print("Lower Quartile (Q1): ", q1)
print("Upper Quartile (Q3): ", q3)
print("Interquartile Range (IQR): ", iqr)
print("Upper Whisker: ", upper_whisker)
print("Lower Whisker: ", lower_whisker)

print(" ")
```

```python
import matplotlib.pyplot as plt

# Create a new figure with a larger size and white background
fig, axs = plt.subplots(3, 1, figsize=(10, 18), facecolor='w')

# Plot 1: MAE Box Plot
data_mae = [mae['LSTM'], mae['CNN'], mae['SCT-CNN-LSTM'], mae['GRU']]
box_colors = ['#6495ED', '#F08080', '#90EE90', '#FFA07A', '#ADD8E6']
median_colors = ['#000000', '#000000', '#000000', '#000000', '#000000']

bp1 = axs[0].boxplot(data_mae, patch_artist=True, labels=['LSTM', 'CNN', 'SCT-CNN-LSTM', 'GRU'])
for patch, color in zip(bp1['boxes'], box_colors):
    patch.set_facecolor(color)
for median, color in zip(bp1['medians'], median_colors):
    median.set_color(color)

axs[0].set_xlabel('Models', fontsize=12)
axs[0].set_ylabel('MAE Score', fontsize=12)
axs[0].set_title('a) MAE Box Plot', fontsize=14)
axs[0].tick_params(axis='x', labelsize=10)
axs[0].tick_params(axis='y', labelsize=10)
axs[0].grid(False)

# Plot 2: MAPE Box Plot
data_mape = [mape['LSTM'], mape['CNN'], mape['SCT-CNN-LSTM'], mape['GRU']]

bp2 = axs[1].boxplot(data_mape, patch_artist=True, labels=['LSTM', 'CNN', 'SCT-CNN-LSTM', 'GRU'])
for patch, color in zip(bp2['boxes'], box_colors):
    patch.set_facecolor(color)
for median, color in zip(bp2['medians'], median_colors):
    median.set_color(color)

axs[1].set_xlabel('Models', fontsize=12)
axs[1].set_ylabel('MAPE Score (%)', fontsize=12)
axs[1].set_title('b) MAPE Box Plot', fontsize=14)
axs[1].tick_params(axis='x', labelsize=10)
axs[1].tick_params(axis='y', labelsize=10)
axs[1].grid(False)

# Plot 3: RMSE Box Plot
data_rmse = [rmse['LSTM'], rmse['CNN'], rmse['SCT-CNN-LSTM'], rmse['GRU']]

bp3 = axs[2].boxplot(data_rmse, patch_artist=True, labels=['LSTM', 'CNN', 'SCT-CNN-LSTM', 'GRU'])
for patch, color in zip(bp3['boxes'], box_colors):
    patch.set_facecolor(color)
for median, color in zip(bp3['medians'], median_colors):
    median.set_color(color)

axs[2].set_xlabel('Models', fontsize=12)
axs[2].set_ylabel('RMSE Score', fontsize=12)
axs[2].set_title('c) RMSE Box Plot', fontsize=14)
axs[2].tick_params(axis='x', labelsize=10)
axs[2].tick_params(axis='y', labelsize=10)
axs[2].grid(False)

# Adjust the spacing between subplots
plt.tight_layout()

#Save the plot as a PNG file

plt.savefig('/Users/musaphiri/Documents/Solar Tracking Research/Solar tracking dataset document/Solar tracking notebooks/Figures/box_plots.png')

#Show the plot

plt.show()
```
